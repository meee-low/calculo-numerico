### Exemplo 1:
Encontrar ra√≠zes de: $f(x) = x^2 -17x +4$

```python
f = lambda x: x**2 - 17*x  + 4

plotar_funcao(f, -2, 20)  # plota de -2 a 20
```

```python
plotar_funcao(f, 0, 1)
```

```python
plotar_funcao(f, 15, 20)
```

D√° para ver que tem ra√≠zes pr√≥ximas de 0.2 e de 17.

#### Pelo m√©todo de Newton:  
$f(x) = x^2 - 17x + 4$  
$f'(x) = 2x - 17$  

```python
derivada = lambda x: 2*x - 17
ger_newton = gerador_newton(f, derivada, x_inicial=17) # cria um gerador que usa o m√©todo de Newton 
usa_gerador(ger_newton, plot=True, digse=4) # encontra uma das ra√≠zes usando esse gerador
```

#### Por aproxima√ß√µes sucessivas

$0 = f(x) = x^2 - 17x + 4$  
$ \Leftrightarrow 17x = x^2 + 4 = 17x $  
$ \Leftrightarrow x = \frac{x^2 + 4}{17}$  

```python
x_isolado = lambda x: (x**2 + 4)/17

ger_aprsuc = gerador_aproximacoes_sucessivas(x_isolado, x_inicial=0.2)
usa_gerador(ger_aprsuc, plot=True)
```
